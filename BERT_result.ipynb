```python
# BigQueryからデータを読み込む
from google.cloud import bigquery
import pandas as pd
import numpy as np
import os
credentials_json = './hypnotic-guard-380905-4d6d5f5cb89c.json'
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = credentials_json
## Set Project ID
# project = 'hypnotic-guard-380905'
## Set Dataset
# dataset = 'tweet_dataset'
## Set Table
# table = 'tweet_table'
bigquery_client = bigquery.Client()
query = 'SELECT * FROM `hypnotic-guard-380905.tweet_dataset.tweet_table`'
rows = bigquery_client.query(query).result()
df = (bigquery_client.query(query)
             .result()
             .to_dataframe(
            create_bqstorage_client=True,
        )
    )
df
```

```python
from transformers import pipeline, AutoModelForSequenceClassification, BertJapaneseTokenizer

dir_name = './model'
model_name = "cl-tohoku/bert-base-japanese-whole-word-masking"
saved_model = AutoModelForSequenceClassification.from_pretrained(dir_name)
tokenizer = BertJapaneseTokenizer.from_pretrained(model_name)

sentiment_analyzer = pipeline("sentiment-analysis", model=saved_model, tokenizer=tokenizer)
```

```python
# 感情分析の実行結果をデータフレームに追加
result = df["Tweet_Text"].apply(lambda x: sentiment_analyzer(x)[0])
df["sentiment"] = result.apply(lambda x: x["label"])
df["score"] = result.apply(lambda x: round(x["score"], 4))
df
```

```python
df = df.drop(["Tweet_ID",'Like','__of_Retweets','Account_ID_','Account_Name','Description','__of_Follow','__of_Follower','Account_Create_Date','Retweet_Tweet_ID','Retweet_Account_ID','Place','Is_Badge'],axis=1)
df
```

```python
#データフレームを BigQueryに保存
client = bigquery.Client(project="hypnotic-guard-380905")
table_id = 'tweet_dataset_result.tweet_table_result'  #予め作成したデータセットIDと作成されるテーブル名を指定
job_config = bigquery.LoadJobConfig(schema=[
    bigquery.SchemaField("Tweet_Text", "STRING", mode="NULLABLE"),
    bigquery.SchemaField("sentiment", "STRING", mode="NULLABLE"),
    bigquery.SchemaField("score", "float64", mode="REQUIRED"),
    bigquery.SchemaField("Tweet_TimeDate_JST_", "STRING", mode="NULLABLE")
])
job = client.load_table_from_dataframe(
    df, table_id, job_config=job_config
)
job.result()
```
